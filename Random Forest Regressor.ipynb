```python
df = pd.read_csv('day_wise.csv', parse_dates=['Date'])
df.head()
```

```python
df.isna().sum()
```

```python
X = df.drop(columns = ['Date', 'Confirmed', 'Deaths'])
y = df['Deaths']
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 64)
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
reg = RandomForestRegressor()
```

```python
reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error 
```

```python
y_pred_train = reg.predict(X_train)
mean_absolute_error(y_train, y_pred_train)
```

```python
y_pred_test = reg.predict(X_test)
mean_absolute_error(y_test, y_pred_test)
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
parameters = {'n_estimators': [100, 120, 130], 
             'max_depth': [10, 20, 30],
              'max_features': [5, 10, 15],
              'max_leaf_nodes': [1, 2, 3]
             }
reg1 = RandomForestRegressor()
reg1 = GridSearchCV(reg1, parameters)
```

```python
reg1.fit(X_train, y_train)
```

```python
y_pred_train = reg1.predict(X_train)
mean_absolute_error(y_train, y_pred_train)
```

```python
y_pred_test = reg1.predict(X_test)
mean_absolute_error(y_test, y_pred_test)
```
